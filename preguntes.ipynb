```python
import pandas as pd

# Ruta de l'arxiu CSV
file_path = '/content/drive/MyDrive/VISUALITZACIÓ DE DADES/PRA 2/dades_2020_2023.csv'

# Llegir l'arxiu CSV amb Pandas
df = pd.read_csv(file_path)

# Mostrar les primeres files del DataFrame per verificar que s'ha llegit correctament
print(df.head())
```

```python
# Assignar la primera fila com a nous noms de columna
new_columns = df.iloc[0]
df.columns = new_columns
# Eliminar la primera fila del DataFrame (opcional, si conté els nous noms de columna)
df = df.iloc[1:].reset_index(drop=True)
```

```python
import pandas as pd

# Suposant que ja tens el DataFrame df creat amb les dades carregades

# Convertir la columna 'Fecha' a tipus datetime si encara no ho és
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Afegir les columnes 'Año' i 'Mes' a partir de 'Fecha'
df['Año'] = df['Fecha'].dt.year
df['Mes'] = df['Fecha'].dt.month
```

```python
import pandas as pd

# Suposem que ja tens el DataFrame carregat i es diu df
# Per exemple, aquí tens com carregar-lo de nou per aquest exemple
# df = pd.read_csv('/ruta/al/fitxer.csv')

# Funció per convertir les columnes a tipus float
def convertir_a_float(df, columnes):
    for col in columnes:
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)
        except ValueError as e:
            print(f"No s'ha pogut convertir la columna {col}: {e}")

    return df

# Columnes que volem convertir a float
columnes_float = [
    'Número de Vuelos', 'Horas Bloque', 'Sillas Ofrecidas',
    'Pasajeros A Bordo', 'Pasajeros Tránsito',
    'Carga Ofrecida (Kg.)', 'Carga + Correo A Bordo (Kg.)', 'Carga Tránsito (Kg.)'
]

# Convertir les columnes a float
df = convertir_a_float(df, columnes_float)

```

```python
df = df.sort_values(by='Fecha')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar per any i mes i sumar els passatgers de trànsit
monthly_data = df.groupby(['Año', 'Mes'], as_index=False)['Pasajeros A Bordo'].sum()

# Crear el gràfic de barres amb seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Mes', y='Pasajeros A Bordo', hue='Año', data=monthly_data, palette='viridis')

plt.title('Evolució del nombre total de passatgers de trànsit per mes i any')
plt.xlabel('Mes')
plt.ylabel('Total passatgers trànsit')
plt.xticks(range(12), ['Gener', 'Febrer', 'Març', 'Abril', 'Maig', 'Juny', 'Juliol', 'Agost', 'Setembre', 'Octubre', 'Novembre', 'Desembre'], rotation=45)
plt.legend(title='Any')

# Mostrar el nombre total de passatgers trànsit en cada barra
for p in plt.gca().patches:
    plt.gca().annotate(f"{int(p.get_height())}", (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Calcular l'ús de la capacitat
df['Capacidad Uso'] = (df['Pasajeros A Bordo'] / df['Sillas Ofrecidas']) * 100

# Filtrar les dades per als anys 2020, 2021 i 2023
df_filtered = df[df['Año'].between(2020, 2023)]

# Crear una funció per generar els gràfics per any
def plot_by_year(year):
    df_year = df_filtered[df_filtered['Año'] == year]
    g = sns.FacetGrid(df_year, col='Mes', col_wrap=3, height=4, aspect=1.5)
    g.map_dataframe(sns.scatterplot, x='Capacidad Uso', y='Pasajeros A Bordo', size='Número de Vuelos', sizes=(20, 200), legend=False)
    g.set_axis_labels('Ús de la capacitat (%)', 'Nombre de passatgers transportats')
    g.set_titles(col_template='Mes {col_name}')
    g.fig.suptitle(f'Relació entre l\'ús de la capacitat dels avions i el nombre de passatgers transportats - Any {year}', y=1.02)
    plt.show()

# Generar els gràfics per a cada any
for year in range(2020, 2024):
    plot_by_year(year)
```

```python
import pandas as pd
import plotly.express as px

# Suposem que ja tens el DataFrame carregat i es diu df
# df = pd.read_csv('/ruta/al/fitxer.csv')

# Convertir la columna 'Fecha' a datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')

# Filtrar les dades per als anys 2020, 2021, 2022 i 2023
df_filtered = df[df['Fecha'].dt.year.between(2020, 2023)]

# Convertir la columna Mes_Any a string per evitar errors de serialització amb plotly
df_filtered['Mes_Any'] = df_filtered['Fecha'].dt.to_period('M').astype(str)

# Agrupar per aerolínia, tipus d'avió, any i mes
aerolines_types_monthly = df_filtered.groupby(['Sigla Empresa', 'Tipo de Equipo', 'Mes_Any']).size().reset_index(name='Num_Vols')

# Crear gràfics interactius amb Plotly per a cada aerolínia
figures = []
aerolines = df_filtered['Sigla Empresa'].unique()

for aerolinea in aerolines:
    data_aerolinea = aerolines_types_monthly[aerolines_types_monthly['Sigla Empresa'] == aerolinea]

    # Crear un gràfic interactiu per cada aerolínia
    fig = px.line(data_aerolinea, x='Mes_Any', y='Num_Vols', color='Tipo de Equipo',
                  title=f'Utilització de diferents tipus d\'avions per {aerolinea}',
                  labels={'Num_Vols': 'Nombre de Vols', 'Mes_Any': 'Mes i Any', 'Tipo de Equipo': 'Tipus d\'avió'})

    fig.update_layout(xaxis_title='Mes i Any', yaxis_title='Nombre de Vols')
    figures.append(fig)

# Mostrar els gràfics interactius
for fig in figures:
    fig.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar les dades (suposant que ja tens el dataframe df amb les dades)
# df = pd.read_csv('ruta/del/arxiu.csv') # Ajusta la ruta segons on tinguis el teu arxiu

# Suposem que df conté les següents columnes: Fecha, Carga Ofrecida (Kg.), Carga + Correo A Bordo (Kg.), Año, Mes

# 1. Filtrar les dades per període (2020-2023) i per variables rellevants
df_filtered = df[(df['Año'] >= 2020) & (df['Año'] <= 2023)]

# 2. Agrupar per any i mes i calcular la suma de càrrega oferida i transportada
df_grouped = df_filtered.groupby(['Año', 'Mes']).agg({
    'Carga Ofrecida (Kg.)': 'sum',
    'Carga + Correo A Bordo (Kg.)': 'sum'
}).reset_index()

# 3. Visualització interactiva amb seaborn o altres llibreries (en aquest cas, utilitzant seaborn)
plt.figure(figsize=(12, 8))

# Plot de la càrrega oferida
sns.lineplot(data=df_grouped, x='Mes', y='Carga Ofrecida (Kg.)', hue='Año', marker='o', markersize=8, linewidth=2)

# Plot de la càrrega transportada
sns.lineplot(data=df_grouped, x='Mes', y='Carga + Correo A Bordo (Kg.)', hue='Año', marker='s', markersize=8, linewidth=2)

plt.title('Variació de la Càrrega Oferta i Transportada per Any i Mes')
plt.xlabel('Mes')
plt.ylabel('Càrrega (Kg)')
plt.grid(True)
plt.legend(title='Any')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar les dades (suposant que ja tens el dataframe df amb les dades)
# df = pd.read_csv('ruta/del/arxiu.csv') # Ajusta la ruta segons on tinguis el teu arxiu

# Suposem que df conté les següents columnes: Continente Origen, Continente Destino, Fecha

# 1. Filtrar i preprocessar les dades
df_filtered = df[(df['Fecha'].dt.year >= 2020) & (df['Fecha'].dt.year <= 2023)]

# 2. Sumaritzar la densitat de vols entre continents
flight_density = df_filtered.groupby(['Continente Origen', 'Continente Destino']).size().unstack(fill_value=0)

# 3. Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(flight_density, annot=True, cmap='YlGnBu', cbar_kws={'label': 'Nombre de Vols'})
plt.title('Densitat de Vols entre Continents (2020-2023)')
plt.xlabel('Continente Destino')
plt.ylabel('Continente Origen')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Suposem que el DataFrame 'df' conté les següents columnes: 'Ciudad Destino', 'Pais Destino'

# Filtrar les dades per excloure destins a Colòmbia
df_exterior = df[df['Pais Destino'] != 'COLOMBIA']

# Comptar el nombre de vols per a cada ciutat de destinació fora de Colòmbia
destinations_count = df_exterior['Ciudad Destino'].value_counts().reset_index()
destinations_count.columns = ['Ciudad Destino', 'Nombre de Vols']

# Convertir les dades a un diccionari per al núvol de paraules
words_dict = dict(zip(destinations_count['Ciudad Destino'], destinations_count['Nombre de Vols']))

# Crear el núvol de paraules
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(words_dict)

# Visualitzar el núvol de paraules
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Ciutats de destinació més freqüents fora de Colòmbia (2020-2023)')
plt.show()
```
