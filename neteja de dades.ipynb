```python
import pandas as pd
import glob

# Ruta a la carpeta amb els fitxers Excel
path = '/content/drive/MyDrive/VISUALITZACIÓ DE DADES/PRA 2'

# Obtenir la llista de tots els fitxers Excel a la carpeta
all_files = glob.glob(path + "/*.xlsx")

# Verificar si s'han trobat fitxers
if not all_files:
    print("No s'han trobat fitxers Excel a la carpeta especificada.")
else:
    # Llista per guardar cada DataFrame
    dfs = []

    # Llegir cada fitxer Excel i afegir-lo a la llista de DataFrames
    for filename in all_files:
        try:
            df = pd.read_excel(filename)
            dfs.append(df)
        except Exception as e:
            print(f"Error en llegir el fitxer {filename}: {e}")

    # Verificar si s'han llegit correctament els fitxers
    if not dfs:
        print("No s'han pogut llegir correctament els fitxers Excel.")
    else:
        # Unir tots els DataFrames en un sol DataFrame
        combined_df = pd.concat(dfs, ignore_index=True)

        # Mostrar les primeres files del DataFrame combinat
        print(combined_df.head())

        # --- Neteja de dades ---

        # Eliminar duplicats
        combined_df.drop_duplicates(inplace=True)

        # Omplir valors nuls per columna segons el tipus de dades
        for column in combined_df.columns:
            if combined_df[column].dtype == 'object':
                combined_df[column].fillna('', inplace=True)  # Omplir textos amb cadena buida
            elif pd.api.types.is_numeric_dtype(combined_df[column]):
                combined_df[column].fillna(combined_df[column].mean(), inplace=True)  # Omplir números amb la mitjana
            elif pd.api.types.is_datetime64_any_dtype(combined_df[column]):
                combined_df[column].fillna(combined_df[column].mode()[0], inplace=True)  # Omplir dates amb la moda

        # Assegurar que les dates es converteixen correctament
        if 'Fecha' in combined_df.columns:
            combined_df['Fecha'] = pd.to_datetime(combined_df['Fecha'], errors='coerce')

            # Crear les columnes 'year' i 'month' a partir de 'Fecha'
            combined_df['year'] = combined_df['Fecha'].dt.year
            combined_df['month'] = combined_df['Fecha'].dt.month

        # Mostrar les primeres files després de la neteja
        print(combined_df.head())

        # Calcular el percentatge de canvi en nombre de passatgers per mes
        if 'passengers' in combined_df.columns:
            combined_df['percent_change_passengers'] = combined_df['passengers'].pct_change() * 100

        # Calcular índex de capacitat d'ús dels avions
        if 'passengers' in combined_df.columns and 'total_seats' in combined_df.columns:
            combined_df['capacity_usage'] = (combined_df['passengers'] / combined_df['total_seats']) * 100

        # Ordenar per data
        if 'Fecha' in combined_df.columns:
            combined_df.sort_values(by='Fecha', inplace=True)

        # Agregar dades per mes si les columnes 'year' i 'month' existeixen
        if 'year' in combined_df.columns and 'month' in combined_df.columns:
            monthly_data = combined_df.groupby(['year', 'month']).agg({
                'flights': 'sum',
                'passengers': 'sum',
                'cargo': 'sum',
                'capacity_usage': 'mean'
            }).reset_index()

            # Guardar dades processades a un nou fitxer CSV
            monthly_data.to_csv('/content/drive/MyDrive/VISUALITZACIÓ DE DADES/PRA 2/dades_processades_2022.csv', index=False)

            # Mostrar les primeres files del DataFrame netejat
            print(monthly_data.head())
        else:
            print("Les columnes 'year' i 'month' no es troben al DataFrame.")

        # Guardar el DataFrame ordenat a un fitxer CSV independentment
        combined_df.to_csv('/content/drive/MyDrive/VISUALITZACIÓ DE DADES/PRA 2/dades_2020_2023.csv', index=False)
        print("Fitxer ordenat guardat.")
```
